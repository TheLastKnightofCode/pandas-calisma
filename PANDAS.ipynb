#Kayıp ve Bozuk Veriler

```python
arr=np.array([["MUSA","OK",np.nan],["AHMET",np.nan,17],["MEHMET","AK",np.nan]])
df=pd.DataFrame(arr,["Index1","Index2","Index3"],["Column1","Column2","Column3"])
df=df.replace("nan",np.nan)
df
```

```python
df.dropna()#NaN olan satırları siler
```

```python
df.dropna(axis=1)#NaN olan sütunları siler
```

```python
df.dropna(thresh=2)#En az 2 tane NaN olmayan veri lazım.Uymayan satırları siler.
```

```python
df.fillna(value=100)
```

```python
df
```

```python
arr2=([12,np.nan,23],[13,45,np.nan],[23,np.nan,np.nan])
df=pd.DataFrame(arr2,["A","B","C"],["D","E","F"])
```

```python
df.sum()
```

```python
df.sum().sum()
```

```python
df.isnull()
```

```python
df.isnull().sum()#hangi column da burda D E F oluyor bunlarda kaç tane NaN olduğunu böyle buluyoruz
```

```python
df.isnull().sum().sum()#toplam kaç tane olduğunu böyle buluyoruz
```

```python
def caluculateMean(df):
  totalSum = df.sum().sum()
  totalNum = df.size - df.isnull().sum().sum()
  return totalSum / totalNum
```

```python
df.fillna(value=caluculateMean(df),inplace=True)#ortalamasını verdik nall değerlerine hep kalsın itiyorsak inplace = True yazmam lazım
```

```python
df
```

#Merge, Join ve Concatenate

```python
dataSet1={
    "A":["A1","A2","A3","A4"],
    "B":["B1","B2","B3","B4"],
    "C":["C1","C2","C3","C4"]
}
dataSet2={
    "A":["A5","A6","A7","A8"],
    "B":["B5","B6","B7","B8"],
    "C":["C5","C6","C7","C8"]
}
df1=pd.DataFrame(dataSet1,index=[1,2,3,4])
df2=pd.DataFrame(dataSet2,index=[5,6,7,8])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
dataSet2={
    "X":["X1","X2","X3"],
    "Y":["Y1","Y2","Y3"]
}
df2=pd.DataFrame(dataSet2,index=[1,2,3])
```

```python
df1.join(df2)#df1 in bütün indexlerini alır ona göre yazar df2 yide.    (JOİNİN how PARAMETRESİ VAR HİÇ BİR ŞEY YAZMAZSAN LEFT YANİ DF1 İN HEPSİNİ,RİGHT DF2 İN HEPSİNİ,OUTER DF1 İLE DF2 NİN HEPSİNİ,İNNER DİYİNCE ORTAK OLAN,İNDEXLERE GÖRE BİRLEŞTİRİR.)
```

```python
df2.join(df1)#df2 nin bütün indexlerini alır ona göre yazar df1 de.
```

```python
dataSet1={
    "A":["A1","A2","A3","A4"],
    "B":["B1","B2","B3","B4"],
    "Anahtar":["K1","K2","K3","K4"]
}
dataSet2={
    "X":["X1","X2","X3"],
    "Y":["Y1","Y2","Y3"],
    "Anahtar":["K1","K2","K3"]
}
df1=pd.DataFrame(dataSet1,index=[1,2,3,4])
df2=pd.DataFrame(dataSet2,index=[1,2,3])
```

```python
pd.merge(df1,df2,on="Anahtar")#joinle aynı mantık bunda column a göre(belirli bir column a göre anahtar column belirlemen lazım) (how parametresi inner olarak ayarlı)
```
